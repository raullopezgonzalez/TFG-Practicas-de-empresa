# Creating dataframe

```python
path = r'C:\Users\riul0\Desktop\TFG_Empresa\Datos_8º_D\csv_8º_D'
files_in_path = os.listdir(path)
files_in_path = files_in_path[2:]
```

```python
for i in range(0,len(files_in_path)):
    csv = pd.read_csv(files_in_path[i]) 
    csv = csv[csv.columns[0]].str.split("\t", n = 1, expand = True) 
    
    zip_interest = zip(csv[0],csv[1])
    columns_of_patient = []
    for j in zip_interest:
        if j[1] == None :
            continue
        else:
            columns_of_patient.append(j[0])
    
    print('Patient : ', files_in_path[i],'with id : ', i,'. . .  Number of columns filled',len(columns_of_patient))
```

```python
patient_with_138_columns = []
patient_with_174_columns = []
for i in range(0,len(files_in_path)):
    csv = pd.read_csv(files_in_path[i]) 
    csv = csv[csv.columns[0]].str.split("\t", n = 1, expand = True)   
    zip_interest = zip(csv[0],csv[1])
    columns_of_patient = []
    for j in zip_interest:
        if j[1] == None :
            continue
        else:
            columns_of_patient.append(j[0])        
    if len(columns_of_patient) == 138:
        patient_with_138_columns.append(i)
    elif len(columns_of_patient) == 174:
        patient_with_174_columns.append(i)
```

```python
print('Patient with 134 columns : ', len(patient_with_138_columns))
print('Patient with 174 columns : ', len(patient_with_174_columns))
print('Total number of patients : ',len(files_in_path))
```
