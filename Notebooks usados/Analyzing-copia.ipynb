## Pairing cycles of patients

```python
print(Dataframe_138_8A_no_zeros_paired.index[0])
print(Dataframe_138_8A_no_zeros_paired.index[68/2])

index_sorted = []
for i in range(0,len(Dataframe_138_8A_no_zeros_paired.index)):
    if i == (int(68/2)):
        break
    else:     
        index_sorted.append(Dataframe_138_8A_no_zeros_paired.index[i])
        index_sorted.append(Dataframe_138_8A_no_zeros_paired.index[int(i + (68/2))])
        
print('')
print(index_sorted)
```

```python
Dataframe_138_8A_no_zeros_paired_sorted = pd.DataFrame(index = index_sorted, columns =  Dataframe_138_8A_no_zeros_paired.columns)
```

```python
Dataframe_138_8A_no_zeros_paired_sorted.head(10)
```

```python
for i in range(0,len(index_sorted)):
    Dataframe_138_8A_no_zeros_paired_sorted.iloc[i] = Dataframe_138_8A_no_zeros_paired.loc[index_sorted[i]]
```
