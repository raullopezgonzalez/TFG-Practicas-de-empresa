```python
comp_means_of_dataframe = []
lista_1_dataframe_means = []
lista_1_dataframe_values = []
lista_2_dataframe_means = []
lista_2_dataframe_values = []

for a in range(0,len(list_of_pairs)):
    index1 = list_of_pairs[a][2]
    index2 = list_of_pairs[a][3]

    
    
    name_of_rows_for_mean = []
    for z in range(0,len(files_in_path[index1][files_in_path[index1].columns[0]])):
        if files_in_path[index1][files_in_path[index1].columns[0]][z] == '-':
            name_of_rows_for_mean.append((files_in_path[index1].index[z], z))


    
    for b in range(0,len(name_of_rows_for_mean)):

        if b == len(name_of_rows_for_mean) - 1:
            lista_1_dataframe_values.append(files_in_path[index1][files_in_path[index1].columns[0]].values[name_of_rows_for_mean[b][1]+1:])
            lista_2_dataframe_values.append(files_in_path[index2][files_in_path[index2].columns[0]].values[name_of_rows_for_mean[b][1]+1:])

        else:
            lista_1_dataframe_values.append(files_in_path[index1][files_in_path[index1].columns[0]].values[name_of_rows_for_mean[b][1]+1:name_of_rows_for_mean[b+1][1]])
            lista_2_dataframe_values.append(files_in_path[index2][files_in_path[index2].columns[0]].values[name_of_rows_for_mean[b][1]+1:name_of_rows_for_mean[b+1][1]])
        
    for c in lista_1_dataframe_values: 
        
        lista_1_dataframe_values_float = list(map(float, c)) 
        mean = np.mean(lista_1_dataframe_values_float)
        lista_1_dataframe_means.append(mean)
        
    for d in lista_2_dataframe_values: 
        lista_2_dataframe_values_float = list(map(float, d)) 
        mean = np.mean(lista_2_dataframe_values_float)
        lista_2_dataframe_means.append(mean)
        
        
    comp_means_of_dataframe.append(a)    
    for e in range(0,len(lista_1_dataframe_means)):
        difference_between_means = lista_1_dataframe_means[e] - lista_2_dataframe_means[e]
        comp_means_of_dataframe.append((name_of_rows_for_mean[e][0],difference_between_means))
    
    
    values = []
    for f in range(1,len(comp_means_of_dataframe)):

        values.append(comp_means_of_dataframe[f][1])
        
    values = [values for values in values if str(values) != 'nan']
        
    #values = values[np.logical_not(np.isnan(values))] 
    
    print('Maximun difference in pair ',list_of_pairs[a])
    print(' -->  ' , np.max(values), 'or ', np.min(values))
    
    print('')

    
    
    #print(comp_means_of_dataframe)
    #print('')
    #print('')
    #print('')
    comp_means_of_dataframe.clear()
    lista_1_dataframe_means.clear()
    lista_1_dataframe_values.clear()
    lista_2_dataframe_means.clear()
    lista_2_dataframe_values.clear()

            
        
      
```
